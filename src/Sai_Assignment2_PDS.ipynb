```python
import pandas as pd

data = pd.read_csv("train.csv")
```

```python
df = pd.DataFrame(data)
null_values = df.isnull().sum()
print(null_values)
```

```python
null_values = df.isnull().sum()
null_values
```

```python

drop_null = df[['Engine', 'Power', 'Seats', 'New_Price']].isnull().all(axis=1)
rows_with_all_empty_data = df[drop_null]
df.drop(rows_with_all_empty_data.index, inplace=True)
```

```python
null_values = df.isnull().sum()
null_values
```

a) Look for the missing values in all the columns and either impute them (replace with mean,
median, or mode) or drop them. Justify your action for this task.

```python

null_values = df.isnull().sum()
df['New_Price'] = df['New_Price'].str.replace(' Lakh', '')
df['New_Price'] = df['New_Price'].str.replace(' Cr', '')
df.dropna(subset=['Mileage'], inplace=True)
print(df['Mileage'].count())
df.dropna(subset=['Seats'], inplace=True)
print(df['Seats'].count())
df.dropna(subset=['New_Price'], inplace=True)
mean_NewPrice = df['New_Price'].astype(float).mean()
df['New_Price'].fillna(mean_NewPrice, inplace=True)
print(df['New_Price'].count())
```

Justification: Using mean of the new_price column might create the bias as there are more number of columns having null values.
There are significant number of rows where the data is missing in more than 3 columns So we have dropped the rows from the data

b) Remove the units from some of the attributes and only keep the numerical values (for
example remove kmpl from “Mileage”, CC from “Engine”, bhp from “Power”, and lakh from
“New_price”).

```python
df['Mileage'] = df['Mileage'].str.replace(' kmpl', '')
df['Mileage'] = df['Mileage'].str.replace(' km/kg', '')
df['Engine'] = df['Engine'].str.replace(' CC', '')
df['Power'] = df['Power'].str.replace(' bhp', '')
df['New_Price'] = df['New_Price'].str.replace(' Lakh', '')
df['New_Price'] = df['New_Price'].str.replace(' Cr', '')
```

e) Perform select, filter, rename, mutate, arrange and summarize with group by operations (or
their equivalent operations in python) on this dataset.

```python
# Select
df_select = df[['Name', 'Year', 'Location']]

print(df_select)
```

```python

# Filter
df_filter = df[(df['Year'] > 2012) & (df['Fuel_Type'] == 'Diesel')]

print(df_filter)
```

```python

# Rename
df_rename = df.rename(columns={'Location': 'City', 'Year': 'Manufacture_Year'})

print(df_rename)
```

```python

df['Mileage'] = pd.to_numeric(df['Mileage'])

# Mutate
df_mutate = df.assign(Kilometers_per_Liter=df['Kilometers_Driven'] / df['Mileage'])

print(df_mutate)
```

```python

# Arrange
df_arranged = df.sort_values(by='Price', ascending=False)

print(df_arranged)
```

```python
# Summarize
df_summary = df.groupby('Transmission').agg({'Year': 'mean', 'Price': 'median'}).rename(columns={'Year': 'Avg_Year', 'Price': 'Medn_Price'}).reset_index()

print(df_summary)
df_summary.head(5)
```
